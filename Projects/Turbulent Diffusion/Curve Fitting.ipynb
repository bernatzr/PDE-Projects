```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import linalg as la
```

```python
def f_model(coef, x):
    y = 0
    for i in range(len(coef)):
        y += coef[i]*x**i
    return y
```

```python
def main(n):
    x = [0.0, 25.0, 50.0, 100.0, 155.0, 200.0]
    y = [0.0, 2.0, 10.0, 50.0, 40.0, 20.0]
    a = np.array([[x[i]**j for j in range(n)]for i in range(len(x))])
    a_t = a.T
    a_t_times_a = np.dot(a_t, a)
    c_hat = np.dot(np.dot(la.inv(a_t_times_a), a_t), y)
    # x_hat = [i for i in range(3, 19)]
    x_hat = np.linspace(0, 200, 100)
    y_hat = [f_model(c_hat, x_i) for x_i in x_hat]
    plt.figure(figsize=(6, 6))
    plt.scatter(x, y)
    plt.plot(x_hat, y_hat)
    # plt.xlabel('x label')
    # plt.ylabel('y label')
    # plt.title("Simple Plot")
    # plt.legend();
    print(c_hat)
    plt.show()
```

```python
main(6)
```
